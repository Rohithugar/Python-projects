# Distribution of the various features in data set and calculate their central tendencies

"Fixed acidity"

```python
plt.figure(figsize = (11,6))

sns.histplot(data = data ,x = 'fixed acidity', color = 'orange', edgecolor = 'linen',
            alpha = 0.5, bins = 5)

plt.title("Histogram of Fixed Acidity")
plt.xlabel('Fixed Acidity')
plt.ylabel('Count')
plt.show()
```

Observation 

We observed that the histogram is normally distributed.
The maximum count of values for fixed acidity lies in between 6 to 8.

```python
round(data['fixed acidity'].mean(),2)
```

```python
data['fixed acidity'].median()
```

Creating the histogram of the "Fixed acidity" feature and also show the mean and the median.

```python
plt.figure(figsize = (11,6))

sns.histplot(data = data ,x = 'fixed acidity', color = 'orange', edgecolor = 'linen',
            alpha = 0.5, bins = 5)

plt.title("Histogram of Fixed Acidity")
plt.xlabel('Fixed Acidity')
plt.ylabel('Count')
plt.vlines(data['fixed acidity'].mean(), ymin=0, ymax = 4000, color = 'blue', label = 'Mean')
plt.vlines(data['fixed acidity'].median(), ymin=0, ymax = 4000, color = 'red', label = 'Median')
plt.legend()
plt.show()
```

Observations

We can see that mean and median are clear representative of the data.

Mean and median are very close to each other.

We can choose either of the parameters say mean as the measure of central tendancy.

Creating the histogram of 'Volatile acidity' feature.

```python
plt.figure(figsize = (11,6))

sns.histplot(data = data ,x = 'volatile acidity', color = 'green', edgecolor = 'linen',
            alpha = 0.5, bins = 5)

plt.title("Histogram of Volatile Acidity")
plt.xlabel('Volatile Acidity')
plt.ylabel('Count')
plt.show()
```

Observation 

We observe that histogram is not well distributed, it is skewed a little towards the right.
As we have seen skewness, therefore we can check the distribution using distplot function.

```python
plt.figure(figsize = (11, 6))
sns.distplot(data['volatile acidity'], color = 'blue')

plt.title("Distplot of Volatile Acidity")
plt.xlabel('Volatile Acidity')
plt.ylabel('Density')
plt.show()
```

Observation

The above plot shows the normal distribution.
The normal distribution is described by the mean and the standard deviation.

the normal distribution is often referred to as a 'bell curve' because of its shape:
    
    - The median and mean are equal
    - It has only one mode
    - It is symmetric, meaning it decreases the same amount on the left and the right of the centre

```python
data['volatile acidity'].skew()
```

skewness value is greater than 1, Hence it is a positively skewed.

```python
data['volatile acidity'].mean()
```

```python
data['volatile acidity'].median()
```

Observation 

The mean and the median are close to each other and the difference between them is very small.
